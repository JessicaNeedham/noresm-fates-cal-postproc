```python
# load libraries needed for plotting
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_size_mort(dbh, ip, r):
    size_mort = 1 / (1 + np.exp(-1 * r * (dbh - ip)))
    return size_mort
```

```python
dbhs = np.linspace(0, 150, 150)
```

```python
print(get_size_mort(60, 140, 0.04))
print(get_size_mort(60, 115, 0.05))
```

```python
fig, axs = plt.subplots(nrows=1,ncols=1,
                        figsize=(10,6))

axs.plot(dbhs, get_size_mort(dbhs, 140, 0.055), label='Current')
axs.plot(dbhs, get_size_mort(dbhs, 134, 0.056), label='Low')
axs.plot(dbhs, get_size_mort(dbhs, 130, 0.057), label='Mid')
axs.plot(dbhs, get_size_mort(dbhs, 125, 0.058), label='High')
axs.legend(fontsize=14)

axs.set_xlabel('DBH (cm)', fontsize=16)
axs.set_ylabel('Size-dependent Mortality Rate', fontsize=16)
axs.set_title('Size-dependent Mortality Function (BETT)', fontsize=18)

# add grid lines
axs.grid(True)
axs.set_xlim(20,80)
axs.set_ylim(0,0.1)

# add a horizontal line at y=0.02
axs.hlines(y=0.02, xmin=20, xmax=80, colors='k', linestyles='--', label='0.02')
```
